```python
#Q3) Split the data in two parts (50%-50%) for training and testing (first half for training, second half for testing). 
#Train a linear regressor model for each variable individually (plus a bias term) and compute the MSE on the training and the testing set. 
#Which variable is the most informative? which one makes the model generalize better? and worse? 
#Compute the coefficient of determination (R^2) for the test set.
```

```python
import numpy as np
from scipy.linalg import lstsq 

loadedData = np.array([map(float, l.split()) for l in open('housing.data')])
trainingData = np.array([map(float, l.split()) for l in open('housing_training.data')])
testingData = np.array([map(float, l.split()) for l in open('housing_testing.data')])

y  = loadedData[:,-1]
y1 = trainingData[:,-1]
y2 = testingData[:,-1]

dot = np.dot #Multiplier
inv = np.linalg.inv #Inversor

#theta_traing = theta_calc(bias,y1)[0]
#theta_test = theta_calc(bias,y2)[0]
half = len(y)/2 #La mitad de las columnas
half = len(y1)/2 #La mitad de las columnas
half = len(y2)/2 #La mitad de las columnas

data_Col_CRIM = 0     # percapita crime rate by town
data_Col_ZN = 1       # proportion of residential land zoned for lots over 25,000 sq.ft.
data_Col_INDUS = 2    # proportion of non-retail business acres per town
data_Col_CHAS = 3     # Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
data_Col_NOX = 4      # nitric oxides concentration (parts per 10 million)
data_Col_RM = 5       # average number of rooms per dwelling
data_Col_AGE = 6      # proportion of owner-occupied units built prior to 1940   
data_Col_DIS = 7      # weighted distances to five Boston employment centres
data_Col_RAD = 8      # index of accessibility to radial highways
data_Col_TAX = 9      # full-value property-tax rate per $10,000
data_Col_PTRATIO = 10 # pupil-teacher ratio by town
data_Col_B = 11       # 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
data_Col_LSTAT = 12   # % lower status of the population
data_Col_MEDV = 13    # Median value of owner-occupied homes in $1000's

print len(y),'data points in housing.data'
print len(y1),'data points in housing_training.data'
print len(y2),'data points in housing_testing.data'

#Definicion de las columnas de training
col_training_CRIM = trainingData[:,data_Col_CRIM:data_Col_CRIM+1]
col_training_ZN = trainingData[:,data_Col_ZN:data_Col_ZN+1]
col_training_INDUS = trainingData[:,data_Col_INDUS:data_Col_INDUS+1]
col_training_CHAS = trainingData[:,data_Col_CHAS:data_Col_CHAS+1]
col_training_NOX = trainingData[:,data_Col_NOX:data_Col_NOX+1]
col_training_RM = trainingData[:,data_Col_RM:data_Col_RM+1]
col_training_AGE = trainingData[:,data_Col_AGE:data_Col_AGE+1]
col_training_DIS = trainingData[:,data_Col_DIS:data_Col_DIS+1]
col_training_RAD = trainingData[:,data_Col_RAD:data_Col_RAD+1]
col_training_TAX = trainingData[:,data_Col_TAX:data_Col_TAX+1]
col_training_PTRATIO = trainingData[:,data_Col_PTRATIO:data_Col_PTRATIO+1]
col_training_B = trainingData[:,data_Col_B:data_Col_B+1]
col_training_LSTAT = trainingData[:,data_Col_LSTAT:data_Col_LSTAT+1]
col_training_MEDV = trainingData[:,data_Col_MEDV:data_Col_MEDV+1]
#Definicion de las columnas de testing
col_testing_CRIM = testingData[:,data_Col_CRIM:data_Col_CRIM+1]
col_testing_ZN = testingData[:,data_Col_ZN:data_Col_ZN+1]
col_testing_INDUS = testingData[:,data_Col_INDUS:data_Col_INDUS+1]
col_testing_CHAS = testingData[:,data_Col_CHAS:data_Col_CHAS+1]
col_testing_NOX = testingData[:,data_Col_NOX:data_Col_NOX+1]
col_testing_RM = testingData[:,data_Col_RM:data_Col_RM+1]
col_testing_AGE = testingData[:,data_Col_AGE:data_Col_AGE+1]
col_testing_DIS = testingData[:,data_Col_DIS:data_Col_DIS+1]
col_testing_RAD = testingData[:,data_Col_RAD:data_Col_RAD+1]
col_testing_TAX = testingData[:,data_Col_TAX:data_Col_TAX+1]
col_testing_PTRATIO = testingData[:,data_Col_PTRATIO:data_Col_PTRATIO+1]
col_testing_B = testingData[:,data_Col_B:data_Col_B+1]
col_testing_LSTAT = testingData[:,data_Col_LSTAT:data_Col_LSTAT+1]
col_testing_MEDV = testingData[:,data_Col_MEDV:data_Col_MEDV+1]
```

```python
def solutionQ3(trainData, testData, columnLabel):
    target = trainingData[:,data_Col_MEDV:data_Col_MEDV+1]
    targetTest = testingData[:,data_Col_MEDV:data_Col_MEDV+1]
    
    # Dummy features / Crea un vector de Ones con la longuitud de Y
    Ones = np.ones((len(target),1))
    
    # Parametros: theta = (X'*X)^-1*X'*y
    Xentreno = np.hstack((Ones, trainData.reshape(len(Ones),1)))
    XTheta = lstsq(Xentreno, target)[0]
    target_pred = dot(Xentreno, XTheta)
    #t = target-target_pred
    #Parameteros: MSE(y,X,theta):
    msePred = sum((target-target_pred)**2)/len(target)
    meanTarget = sum(target)/len(target)
    varianceTarget = sum((target-meanTarget)**2)/len(target)
    Coef = msePred/varianceTarget
    
    Xtest = np.hstack((Ones, testData.reshape(len(Ones),1)))
    #theta conjunto de entrenamiento
    target_pred_test = dot(Xtest, XTheta)
    
    msePredTest = sum((targetTest-target_pred_test)**2)/len(targetTest)
    meanTargetTest = sum(targetTest)/len(targetTest)
    varianceTargetTest = sum((targetTest-meanTargetTest)**2)/len(targetTest)
    CoefTest = msePredTest/varianceTargetTest

    #print '###',columnLabel,'###'
    print 'MSE training set:', msePred
    print 'MSE testing set:', msePredTest
    print 'R2  testing set theta training set:', 1 - CoefTest,'\n'

    
solutionQ3(col_training_CRIM, col_testing_CRIM, "CRIM")
solutionQ3(col_training_ZN, col_testing_ZN, "ZN")
solutionQ3(col_training_INDUS, col_testing_INDUS, "INDUS")
solutionQ3(col_training_CHAS, col_testing_CHAS, "CHAS")
solutionQ3(col_training_NOX, col_testing_NOX, "NOX")
solutionQ3(col_training_RM, col_testing_RM, "RM")
solutionQ3(col_training_AGE, col_testing_AGE, "AGE")
solutionQ3(col_training_DIS, col_testing_DIS, "DIS")
solutionQ3(col_training_RAD, col_testing_RAD, "RAD")
solutionQ3(col_training_TAX, col_testing_TAX, "TAX")
solutionQ3(col_training_PTRATIO, col_testing_PTRATIO, "PTRATIO")
solutionQ3(col_training_B, col_testing_B, "B")
solutionQ3(col_training_LSTAT, col_testing_LSTAT, "LSTAT")
```
